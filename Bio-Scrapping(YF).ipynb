# Load The SYMBOLS Data Set

```python
Symbols=pd.read_csv("Symbols.csv")
```

# Converting The Array Value To A Single List

```python
tickers_list=[]
for t in list(Symbols.values):
    print(t)
    x=t[0]
    tickers_list.append(x)
```

# Printing The Company Name For Each Of The Symbol / Stock

```python
for t in list(Symbols.values):
    print(t)
    x=t[0]
    company_name = yf.Ticker(x).info['longName']
    print(company_name)
```

# Downloading All The Stocks / Symbols Data 

I got 1230 columns while downloading all the symbols/stocks

# Filtering The Dataset 

We want only last five years data (From 2017-02-01 to 2022-02-22)

```python
Data=pd.DataFrame(Data)
Data.sort_index(ascending=False)
```

```python
Data.shape
```

```python
Data=Data.loc[Data.index>'2017-02-01']
Data=pd.DataFrame(Data)
```

# Flattening The Datasets From Multi-Index

```python
df=pd.DataFrame(Data.to_records())
df.head(5)
```
